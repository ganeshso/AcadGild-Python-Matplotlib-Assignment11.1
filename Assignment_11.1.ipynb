```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
```

## 1. Create a pie chart presenting the male/female proportion 

```python
#titanic.info()
#titanic.sex.fillna('male')
titanic.sex.value_counts(normalize=True)
```

```python
plt.pie(titanic.sex.value_counts(),labels=['Male','Female'],radius=1.5,shadow=True,autopct='%1.2f%%')
plt.show()
```


## 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
import numpy as np
farepaid  = titanic['fare']
age = titanic['age']
clr = np.where(titanic.sex =='female','yellow','red')
plt.title("Scatterplot with the Fare paid and the Age, differ the plot color by gender", fontsize=10)
plt.scatter(farepaid,age,c=clr,alpha=0.9,s=100)
#plt.xlabel(titanic.age.name.title(),fontsize=20)
plt.ylabel("Age",fontsize=20)
plt.xlabel("Fare Paid", fontsize=20)
plt.legend(loc=1)
plt.show()
```

```python
plt.figure(figsize=(22,8))  # create a plot figure

Male_age = titanic[titanic.sex=='male'].age
Male_fare = titanic[titanic.sex=='male'].fare

Female_age = titanic[titanic.sex=='female'].age
Female_fare = titanic[titanic.sex=='female'].fare

#clr = titanic.sex.unique()
size = titanic.fare 
plt.title("Scatter plot with the Fare paid and the Age, differ the plot color by gender",fontsize=20)
plt.scatter(Male_age,Male_fare,s=size,marker='o',label="Male")

plt.scatter(Female_age,Female_fare,s=size,marker='v' ,label="Female")
plt.xlabel(titanic.age.name.title(),fontsize=30)
plt.ylabel(titanic.fare.name.title(), fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(handlelength=2, borderpad=1.25, labelspacing=1,fontsize=20)
plt.show()
```
